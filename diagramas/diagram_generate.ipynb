```python
#QSVM
from qiskit.circuit.library import ZZFeatureMap
import matplotlib.pyplot as plt


feature_map = ZZFeatureMap(feature_dimension=2, reps=1) 

# Decompõe para mostrar as portas H e CNOTs explicitamente
feature_map_decomposed = feature_map.decompose()

feature_map_decomposed.draw(output='mpl', filename='diagrama_zzfeaturemap.png')
print("imagem salva")
```

```python
#QNN

from qiskit.circuit.library import RealAmplitudes
import matplotlib.pyplot as plt


# reps=1 mostra uma camada de rotação + entrelaçamento + rotação
ansatz = RealAmplitudes(num_qubits=2, reps=1)

# Decompõe para mostrar as portas Ry (Rotação Y) e os CNOTs
ansatz_decomposed = ansatz.decompose()


ansatz_decomposed.draw(output='mpl', filename='diagrama_real_amplitudes.png')
print("imagem salva")
```
